# 🧠 Recursive Insight Agent Notebook
This notebook processes ChatGPT `.json` exports into recursive insights using a local LLM stub.

It includes:
- ✅ Load ChatGPT `.json` exports
- ✅ Extract messages from conversations
- ✅ Filter and format insights
- ✅ Send prompts to `query_llama()` (stub)
- ✅ Save outputs as markdown or use in agents

```python
# 📦 Imports
import json
from pathlib import Path
from typing import List
from pprint import pprint
```

```python
# 📁 Config
CHATGPT_JSON_PATH = Path("C:/Users/ANN/Hub/conversations.json")
```

```python
# 📥 Load & Extract Messages
def load_exported_conversations(path: Path) -> List[str]:
    with open(path, 'r', encoding='utf-8') as f:
        all_convos = json.load(f)
    messages = []
    for convo in all_convos:
        mapping = convo.get("mapping", {})
        for node in mapping.values():
            msg_obj = node.get("message", {})
            content = msg_obj.get("content", {}).get("parts", [])
            if isinstance(content, list):
                messages.extend(content)
            elif isinstance(content, str):
                messages.append(content)
    return messages

messages = load_exported_conversations(CHATGPT_JSON_PATH)
pprint(messages[:5])
```

```python
# 🧠 Extract Recursive Insights
def extract_insights(messages: List[str]) -> List[str]:
    insights = []
    for msg in messages:
        if len(msg) > 50:
            insights.append("🧠 " + msg[:200] + ("..." if len(msg) > 200 else ""))
    return insights

insights = extract_insights(messages)
pprint(insights[:5])
```

```python
# 🔁 Stub for Local LLM (Replace with actual model call)
def query_llama(prompt: str) -> str:
    return f"[LLM Insight Placeholder]: {prompt[:100]}..."

llm_responses = [query_llama(i) for i in insights[:5]]
pprint(llm_responses)
```
