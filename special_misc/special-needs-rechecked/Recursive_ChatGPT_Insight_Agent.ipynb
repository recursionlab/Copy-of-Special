# 🤖 Recursive ChatGPT Insight Agent
Turns exported `.json` ChatGPT conversations into structured insights with `query_llama()` integration.


```python
# 📦 Imports
import json
from pathlib import Path
from typing import List, Dict
from pprint import pprint
```

```python
# 📁 Config
CHATGPT_JSON_PATH = "C:/Users/ANN/Hub/conversations.json"
```

```python
def load_exported_conversations(path: str) -> List[str]:
    with open(path, 'r', encoding='utf-8') as f:
        all_convos = json.load(f)

    messages = []
    for convo in all_convos:
        mapping = convo.get("mapping", {})
        for node in mapping.values():
            msg_obj = node.get("message", {})
            content = msg_obj.get("content", {}).get("parts", [])
            if isinstance(content, list):
                messages.extend(content)
            elif isinstance(content, str):
                messages.append(content)

    return messages
```

```python
# 🔍 Recursive Insight Extractor
def extract_insights(messages: List[str]) -> List[str]:
    insights = []
    for msg in messages:
        if len(msg) > 50:
            insights.append("🧠 " + msg[:200] + ("..." if len(msg) > 200 else ""))
    return insights
messages = load_exported_conversations(CHATGPT_JSON_PATH)
insights = extract_insights(messages)
pprint(insights[:5])
```

```python
# 🧠 Stub: query_llama — replace this with your local model interface
def query_llama(prompt: str) -> str:
    return f"[LLM Insight Placeholder]: {prompt[:100]}..."

# 🌀 Run model on extracted insights
llm_responses = [query_llama(i) for i in insights[:5]]  # limit to 5 for now
pprint(llm_responses)
```
