```python

# 🧠 Recursive Insight Agent — Full Notebook

# 📦 Imports
import json
from pathlib import Path
from typing import List
from pprint import pprint
from IPython.display import Markdown, display

# 📁 Configuration
CHATGPT_JSON_PATH = Path("C:/Users/ANN/Hub/conversations.json")

# 🧱 Define Recursive Insight Agent
class RecursiveInsight:
    def __init__(self, prompt: str, answer: str):
        self.prompt = prompt.strip()
        self.answer = answer.strip()

    def extract_summary(self) -> str:
        return f"🧠 **Prompt:** {self.prompt[:60]}...\n**Insight:** {self.answer[:200]}..."

    def to_markdown(self) -> 'Markdown':
        return Markdown(self.extract_summary())

    def __repr__(self) -> str:
        return self.extract_summary()

# 🧹 Load + Clean + Distill ChatGPT JSON Export
def load_and_distill(path: Path) -> List[RecursiveInsight]:
    with open(path, 'r', encoding='utf-8') as f:
        all_convos = json.load(f)

    insights = []
    for convo in all_convos:
        mapping = convo.get("mapping", {})
        prompt = None
        for node in mapping.values():
            msg = node.get("message")
            if not isinstance(msg, dict):
                continue

            author = msg.get("author", {})
            if not isinstance(author, dict):
                continue
            role = author.get("role", "")

            content = msg.get("content", {})
            if not isinstance(content, dict):
                continue
            parts = content.get("parts", [])

            if not isinstance(parts, list) or not parts:
                continue

            raw_part = parts[0]
            text = raw_part.strip() if isinstance(raw_part, str) else raw_part.get("text", "").strip()

            if not text:
                continue

            if role == "user":
                prompt = text
            elif role == "assistant" and prompt:
                insights.append(RecursiveInsight(prompt, text))
                prompt = None

    return insights

# ✅ Ensure insights are loaded first
insights = load_and_distill(CHATGPT_JSON_PATH)

# 🧠 Stub: Replace with your local LLaMA interface
def query_llama(prompt: str) -> str:
    return f"[LLM Insight Placeholder]: {prompt[:100]}..."

# 🚀 Run stub model
llm_outputs = [query_llama(i.answer) for i in insights[:5]]
pprint(llm_outputs)

# 🪄 Show Insights in Markdown
for insight in insights[:5]:
    display(insight.to_markdown())

# 📥 Join Full Conversations Before Extracting
def join_conversation_threads(path: Path) -> List[RecursiveInsight]:
    with open(path, 'r', encoding='utf-8') as f:
        all_convos = json.load(f)

    insights = []
    for convo in all_convos:
        mapping = convo.get("mapping", {})
        convo_thread = []
        for node in mapping.values():
            msg = node.get("message", {})
            if not isinstance(msg, dict): continue

            author = msg.get("author", {}).get("role", "")
            parts = msg.get("content", {}).get("parts", [])
            if not isinstance(parts, list) or not parts: continue

            raw_part = parts[0]
            text = raw_part.strip() if isinstance(raw_part, str) else raw_part.get("text", "").strip()
            if not text: continue

            convo_thread.append((author, text))

        user_blocks = [x[1] for x in convo_thread if x[0] == "user"]
        assistant_blocks = [x[1] for x in convo_thread if x[0] == "assistant"]
        if user_blocks and assistant_blocks:
            prompt = "\n\n".join(user_blocks)
            answer = "\n\n".join(assistant_blocks)
            insights.append(RecursiveInsight(prompt, answer))

    return insights

# 📤 Export insights to Markdown
def export_insights_md(insights: List[RecursiveInsight], out_path="insights.md"):
    with open(out_path, 'w', encoding='utf-8') as f:
        for i, insight in enumerate(insights, 1):
            f.write(f"## 🧠 Insight {i}\n")
            f.write(f"### Prompt\n{insight.prompt}\n\n")
            f.write(f"### Answer\n{insight.answer}\n\n")
            f.write("---\n\n")
```
