# 🧠 Recursive Insight Agent Notebook
This notebook processes ChatGPT JSON exports into recursive insights using a local LLM stub. PDF support can be added later modularly.


```python
# 📦 Imports
import json
from pathlib import Path
from typing import List
from pprint import pprint
```

```python
# 📁 Path to exported ChatGPT JSON
CHATGPT_JSON_PATH = Path("C:/Users/ANN/Hub/conversations.json")
```

```python
# 🧱 Load JSON safely
def load_exported_conversations(path: Path) -> List[str]:
    with open(path, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            print("❌ Failed to parse JSON.")
            return []
    
    messages = []
    for convo in data:
        mapping = convo.get("mapping")
        if not mapping:
            continue
        for node in mapping.values():
            msg_obj = node.get("message") or {}
            content = msg_obj.get("content", {}).get("parts", [])
            if isinstance(content, list):
                messages.extend(content)
            elif isinstance(content, str):
                messages.append(content)
    return messages
```

```python
# 🔍 Load and preview messages
messages = load_exported_conversations(CHATGPT_JSON_PATH)
pprint(messages[:5])
```

```python
# 🧠 Insight Extractor
def extract_insights(messages: List[str]) -> List[str]:
    insights = []
    for msg in messages:
        if isinstance(msg, str) and len(msg) > 50:
            clean_msg = msg.strip().replace('\n', ' ')
            insights.append("🧠 " + clean_msg[:200] + ("..." if len(clean_msg) > 200 else ""))
    return insights
```

```python
# 🔍 Extract and view insights
insights = extract_insights(messages)
pprint(insights[:5])
```

```python
# ⚙️ Stub for local LLM query
def query_llama(prompt: str) -> str:
    return f"[LLM Insight Placeholder]: {prompt[:100]}..." 
```

```python
# 🌀 Run LLM on insights
llm_responses = [query_llama(i) for i in insights[:5]]
pprint(llm_responses)
```
