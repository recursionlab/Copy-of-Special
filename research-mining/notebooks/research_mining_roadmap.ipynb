````python
# Define ConceptUnit Schema
from typing import List, Optional
from datetime import datetime
from pydantic import BaseModel, Field

class ConceptUnit(BaseModel):
    id: str = Field(..., pattern=r'^[a-zA-Z0-9_-]+$')
    version: Optional[str] = "1.0.0"
    source: str
    type: str  # Could use Literal for enum
    definition: str = Field(..., min_length=10)
    mathematical_formulation: Optional[str] = None
    relations: Optional[List[str]] = []
    examples: Optional[List[str]] = []
    prototypes: Optional[List[str]] = []
    references: Optional[List[str]] = []
    tags: Optional[List[str]] = []
    extraction_date: Optional[str] = None
    confidence_score: Optional[float] = Field(None, ge=0, le=1)
    notes: Optional[str] = None

# Enhanced Example ConceptUnit
example_unit = ConceptUnit(
    id="functor-001",
    version="1.0.0",
    source="Thinking with Functors",
    type="Functor",
    definition="A functor is a mapping between categories that preserves the structure of morphisms and objects.",
    mathematical_formulation="F: C → D, F(f: A→B) = F(f): F(A)→F(B)",
    relations=["recursion", "morphism", "category"],
    examples=["List<T> -> Set<T>", "Maybe<T> -> Either<Error, T>"],
    prototypes=["prototypes/functor_graph_demo.py"],
    references=["Mac Lane, S. (1998). Categories for the Working Mathematician."],
    tags=["category theory", "structure preservation", "mapping"],
    extraction_date=datetime.now().isoformat(),
    confidence_score=0.95,
    notes="Fundamental concept in category theory with applications in functional programming."
)

print("Enhanced ConceptUnit created:")
print(example_unit.json(indent=2))
````

````python
# 2. Define ConceptUnit Schema

Define the enhanced JSON schema for ConceptUnits using Python dictionaries or classes, including validation and TypeScript interface.

## Enhanced Schema Features:
- Versioning for units
- Mathematical formulations
- References and tags
- Extraction metadata (date, confidence)
- Validation rules (patterns, minLength, enums)

## JSON Schema (v0.2):
```json
{
  "$schema": "http://json-schema.org/draft-07/schema#",
  "id": "concept-unit-schema",
  "version": "0.2",
  "description": "Enhanced schema for ConceptUnits in research-mining",
  "type": "object",
  "properties": {
    "id": {"type": "string", "pattern": "^[a-zA-Z0-9_-]+$"},
    "version": {"type": "string", "default": "1.0.0"},
    "source": {"type": "string"},
    "type": {"type": "string", "enum": ["Functor", "Recursion", "Autopoiesis", "Distinction", "Contradiction"]},
    "definition": {"type": "string", "minLength": 10},
    "mathematical_formulation": {"type": "string"},
    "relations": {"type": "array", "items": {"type": "string"}, "uniqueItems": true},
    "examples": {"type": "array", "items": {"type": "string"}},
    "prototypes": {"type": "array", "items": {"type": "string"}},
    "references": {"type": "array", "items": {"type": "string"}},
    "tags": {"type": "array", "items": {"type": "string"}},
    "extraction_date": {"type": "string", "format": "date-time"},
    "confidence_score": {"type": "number", "minimum": 0, "maximum": 1},
    "notes": {"type": "string"}
  },
  "required": ["id", "source", "type", "definition"],
  "additionalProperties": false
}
```

## TypeScript Interface:
```typescript
interface ConceptUnit {
  id: string;
  version?: string;
  source: string;
  type: 'Functor' | 'Recursion' | 'Autopoiesis' | 'Distinction' | 'Contradiction';
  definition: string;
  mathematical_formulation?: string;
  relations?: string[];
  examples?: string[];
  prototypes?: string[];
  references?: string[];
  tags?: string[];
  extraction_date?: string;
  confidence_score?: number;
  notes?: string;
}
```
````
