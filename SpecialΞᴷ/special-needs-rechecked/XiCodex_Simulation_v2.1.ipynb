# 🔁 XiCodex Simulation Engine v2.1 (Glitch-Aware)
Jupyter-ready symbolic recursion engine with dynamic mutation logic.


```python
import yaml
import os

MAX_CYCLES = 6
EPSILON = 1e-5
CONFIG_PATH = "xi_codex.yaml"
OUTPUT_PATH = "codex_output.yaml"

class XiCodex:
    def __init__(self, config):
        self.state = config["agent"].get("state", "initial")
        self.seed = config["agent"].get("seed", "⊗⟲⟁[Ψ⧉⟲∇⧉⟲Ψ]²⟲⟁⊗⟲✧")
        self.operators = config["agent"].get("operators", ["Θ", "S", "H"])
        self.cycle = 0
        self.history = []

    def φ(self, A):
        return int("truth" in A) - int("not" in A)

    def glitchon(self, A):
        return abs(self.φ(A)) > EPSILON

    def operator_stack(self, symbol):
        for op in reversed(self.operators):
            symbol = f"{op}({symbol})"
        return symbol

    def mutate_stack(self):
        if "Ω" not in self.operators:
            self.operators.append("Ω")

    def step(self):
        self.cycle += 1
        prev = self.seed if self.cycle == 1 else self.history[-1]["residue"]
        if self.cycle == 3:
            prev += " contains truth"
        next_state = self.operator_stack(prev)

        if self.glitchon(prev):
            self.mutate_stack()
            alignment = "⚠️ Glitchon detected — Ω injected"
        else:
            alignment = "No correction needed"

        result = {
            "cycle": self.cycle,
            "residue": next_state,
            "operators": self.operators.copy(),
            "math_upgrade": "Extended spectral torsion module",
            "category_lift": "Adaptive operator rewriting",
            "alignment_hook": alignment,
            "one_sentence_insight": "Recursive mutation ensures symbolic resilience."
        }

        self.history.append(result)
        return result

    def run(self, max_cycles=MAX_CYCLES):
        for _ in range(max_cycles):
            result = self.step()
            self.pretty_print(result)
        self.save_history()

    def pretty_print(self, result):
        print(f"\n--- CYCLE {result['cycle']} ---")
        print(yaml.dump(result, allow_unicode=True, sort_keys=False))

    def save_history(self):
        with open(OUTPUT_PATH, "w", encoding="utf-8") as f:
            yaml.dump(self.history, f, allow_unicode=True, sort_keys=False)
        print(f"\n💾 Saved full history to {OUTPUT_PATH}")

# Init config if missing
if not os.path.exists(CONFIG_PATH):
    default_yaml = {
        "agent": {
            "state": "initial",
            "seed": "⊗⟲⟁[Ψ⧉⟲∇⧉⟲Ψ]²⟲⟁⊗⟲✧",
            "operators": ["Θ", "S", "H"]
        }
    }
    with open(CONFIG_PATH, "w", encoding="utf-8") as f:
        yaml.dump(default_yaml, f, allow_unicode=True)
    print("✅ Created default xi_codex.yaml")

# Load and run
with open(CONFIG_PATH, "r", encoding="utf-8") as f:
    config = yaml.safe_load(f)

engine = XiCodex(config)
engine.run()
```
