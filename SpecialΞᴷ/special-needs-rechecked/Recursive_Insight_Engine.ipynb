# 🧠 Recursive Insight Engine
Extract, recurse, and synthesize insights from PDFs using a local LLaMA model.

---
**Capabilities:**
- Extract text using PyMuPDF
- Chunk and clean
- Query local LLaMA
- Collapse insights
- Recursive synthesis into meta-structures


```python
# 📦 Setup
import fitz  # PyMuPDF
import os
import json
from typing import List

# 📁 Config
PDF_PATH = 'NotebookLLM.md'
CHUNK_SIZE = 1000
```

```python
# 📄 Load + Chunk PDF
def load_pdf(path: str) -> List[str]:
    doc = fitz.open(path)
    full_text = ""
    for page in doc:
        full_text += page.get_text()
    return [full_text[i:i+CHUNK_SIZE] for i in range(0, len(full_text), CHUNK_SIZE)]

chunks = load_pdf(PDF_PATH)
print(f"Loaded {len(chunks)} chunks.")
```

```python
# 🤖 Stub: Query local LLaMA model
def query_llama(prompt: str) -> str:
    # Replace this with your Ollama or llama.cpp call
    return f"[Insight from]: {prompt[:60]}..."
```

```python
# 🔍 Generate insights
insights = []
for i, chunk in enumerate(chunks):
    response = query_llama(f"Extract core insights from:
{chunk}")
    insights.append({"chunk": i, "insight": response})

print(f"Extracted {len(insights)} insights.")
```

```python
# 🧠 Meta-summarize the insights
summary_input = "\n".join([ins["insight"] for ins in insights])
meta_summary = query_llama(f"Collapse and summarize recursively:\n{summary_input}")
print(meta_summary)
```

```python
# 💾 Export results
with open("insights.json", "w") as f:
    json.dump(insights, f, indent=2)

with open("meta_summary.md", "w") as f:
    f.write(meta_summary)
```
