```python
# 📦 Imports 
import json
from pathlib import Path
from typing import List
from pprint import pprint
from IPython.display import Markdown, display
```

```python
# 📁 Configuration 
CHATGPT_JSON_PATH = Path("C:/Users/ANN/Hub/conversations.json")

```

```python
# 🧱 Define Recursive Insight Agent 
class RecursiveInsight:
    def __init__(self, prompt: str, answer: str):
        self.prompt = prompt.strip()
        self.answer = answer.strip()

    def extract_summary(self) -> str:
        return f"🧠 **Prompt:** {self.prompt[:60]}...\n**Insight:** {self.answer[:200]}..."

    def to_markdown(self) -> 'Markdown':
        return Markdown(self.extract_summary())

    def __repr__(self) -> str:
        return self.extract_summary()
```

```python
# 🧹 Load + Clean + Distill ChatGPT JSON Export
def load_and_distill(path: Path) -> List[RecursiveInsight]:
    with open(path, 'r', encoding='utf-8') as f:
        all_convos = json.load(f)

    insights = []
    for convo in all_convos:
        mapping = convo.get("mapping", {})
        prompt = None
        for node in mapping.values():
            msg = node.get("message")
            if not isinstance(msg, dict):
                continue

            author = msg.get("author", {})
            if not isinstance(author, dict):
                continue
            role = author.get("role", "")

            content = msg.get("content", {})
            if not isinstance(content, dict):
                continue
            parts = content.get("parts", [])

            if not isinstance(parts, list) or not parts:
                continue

            # ✅ Handle str or dict parts defensively
            raw_part = parts[0]
            text = raw_part.strip() if isinstance(raw_part, str) else raw_part.get("text", "").strip()

            if not text:
                continue

            if role == "user":
                prompt = text
            elif role == "assistant" and prompt:
                insights.append(RecursiveInsight(prompt, text))
                prompt = None

    return insights


```

```python
# ✅ Ensure insights are loaded first
insights = load_and_distill(CHATGPT_JSON_PATH)

# 🧠 Stub: Replace with your local LLaMA interface
def query_llama(prompt: str) -> str:
    return f"[LLM Insight Placeholder]: {prompt[:100]}..."

# 🚀 Run stub model
llm_outputs = [query_llama(i.answer) for i in insights[:5]]
pprint(llm_outputs)
```

```python
# 🪄 Show Insights in Markdown
for insight in insights[:5]:
    display(insight.to_markdown())
```
